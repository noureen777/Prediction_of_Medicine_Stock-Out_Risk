```python
import pandas as pd
import re
```

```python
# load sata
df = pd.read_csv('text_data_unclean.csv')
```

```python
# cleaning function
def clean_text(text):
    text = str(text).lower()          # lowercase
    text = text.replace('3', 'e')     # fix '3' -> 'e'
    text = text.replace('@', 'a')     # fix '@' -> 'a'
    text = text.replace('?', '')      # remove '?'
    text = text.replace(' ...', '')   # fix example like: 'dr ...ugs' -> 'drugs'
    text = re.sub(r'\s+', ' ', text)  # remove extra spaces
    return text.strip()
```

```python
df['text_clean'] = df['text'].apply(clean_text)
```

```python
df.head()
```

```python
df = df.drop(columns=['id'])
```

**BERT is SMARTER** -> BERT doesn't need a separate column to tell it the "Sentiment"; it reads the sentence and feels the sentiment itself.

**BERT is DEEP** -> BERT doesn't need you to count keywords it understands that "queue" + "import" implies a logistics failure, which is much more powerful than just knowing "Count = 2"

```python
#therefore i'm going to drop [keyword_count, sentiment, risk_score]

# and also i don't want to limit the model to my calculated features
# i wanted it to learn the true features from the source
```
